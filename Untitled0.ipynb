```python
import pandas as pd
from arch import arch_model
import yfinance as yf
```

```python
data = yf.download("^NSEI", period="1y")["Close"].dropna()
print(data)
```

```python
returns = 100 * data.pct_change().dropna()
print(returns)
```

```python
model = arch_model(returns, vol="Garch", p=1, q=1)
res = model.fit(disp="off")
print(res.summary())
```
